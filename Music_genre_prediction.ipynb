# EDA

# X and y

```python
X = df.drop(columns = ["filename", "label"])
y = df["label"]
```

# Train test split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

# Preprocessing pipeline

```python
from sklearn.preprocessing import StandardScaler
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline

num_cols = X.columns
num_trans = StandardScaler()

preprocessor = make_column_transformer((num_trans, num_cols))
```

# Creating a pipeline

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
```

```python
classifiers = [
    LogisticRegression(max_iter = 1000),
    DecisionTreeClassifier(),
    RandomForestClassifier(n_jobs = -1),
    SVC(),
    KNeighborsClassifier(n_jobs = -1)
]

for classifier in classifiers:
    pipe = make_pipeline(preprocessor, classifier)
    grid = GridSearchCV(estimator = pipe, cv = 5, param_grid = {}, scoring = 'accuracy')
    grid.fit(X_train, y_train)
    
    print(f"{classifier}: {grid.best_score_}")
```

```python
# Choosing to do Hyperparameter tuning for Randomforest, SVC and KNeighbors
```

# Hyperparameter tuning

```python
# KNeighbors
param_grid_knn = {'kneighborsclassifier__n_neighbors': [2, 5, 10],
                 'kneighborsclassifier__p': [1, 2],
                 'kneighborsclassifier__weights': ['uniform', 'distance']}

#SVC
param_grid_svc = {'svc__gamma': ['auto', 'scale'],
                 'svc__C':[0.1, 1, 3, 10]}

#RandomForestClassifier
param_grid_random_forest = {'randomforestclassifier__n_estimators': [50, 100, 200, 500],
                           'randomforestclassifier__max_depth': [None, 5, 10, 20]}

# Models
knn = KNeighborsClassifier(n_jobs = -1)
svc = SVC()
random_forest = RandomForestClassifier(n_jobs = -1)

# Pipes
pipe_knn = make_pipeline(preprocessor, knn)
pipe_svc = make_pipeline(preprocessor, svc)
pipe_random_forest = make_pipeline(random_forest)

# Tuning for each model
grid_knn = GridSearchCV(estimator = pipe_knn, cv = 5, param_grid = param_grid_knn, scoring = "accuracy")
grid_svc = GridSearchCV(estimator = pipe_svc, cv = 5, param_grid = param_grid_svc, scoring = "accuracy")
grid_random_forest = GridSearchCV(estimator = pipe_random_forest, cv = 5, param_grid = param_grid_random_forest, scoring = "accuracy")

grid_knn.fit(X_train, y_train)
grid_svc.fit(X_train, y_train)
grid_random_forest.fit(X_train, y_train)

print(f"{grid_knn.best_score_}, best params: {grid_knn.best_params_}")
print(f"{grid_svc.best_score_}, best params: {grid_svc.best_params_}")
print(f"{grid_random_forest.best_score_}, best params: {grid_random_forest.best_params_}")
```

```python
# The best model is KNN, n_neighbors = 2, p = 1 and weights = 'distance'
```

# Predictions on test data

```python
model = grid_knn.best_estimator_
```

```python
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score

y_pred = model.predict(X_test)
y_true = y_test

accuracy = accuracy_score(y_true, y_pred)

print(f"Accuracy: {accuracy}")
```
